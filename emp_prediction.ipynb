```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import joblib
from lightgbm import LGBMClassifier
```

```python
df=pd.read_csv("BankChurners.csv")
```

```python
df.drop(columns=["Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2"],inplace=True)
```

```python
df.drop(columns=["Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1"],inplace=True)
```

```python
df.drop(columns=["CLIENTNUM"],inplace=True)
```

```python
df
```

```python
categorical=df.select_dtypes(exclude=['int64','float64'])
```

```python
for var in categorical.columns:
    print(categorical[var].value_counts())  
```

```python
df["Attrition_Flag"]=df["Attrition_Flag"].map({"Existing Customer":0,"Attrited Customer":1})
```

```python
df["Gender"]=df["Gender"].map({"M":0,"F":1})
```

```python
df["Marital_Status"]=df["Marital_Status"].map({"Married":3,"Single":2,"Divorced":1,"Unknown":0})
```

```python
oe=OrdinalEncoder(categories=[["Doctorate","Post-Graduate","Graduate","College","High School","Uneducated","Unknown"]])
```

```python
pp=oe.fit_transform(df[["Education_Level"]])
```

```python
df["Education_Level"]=pp
```

```python
oei=OrdinalEncoder(categories=[["$120K +","$80K - $120K","$60K - $80K","$40K - $60K","Less than $40K","Unknown"]])
```

```python
df["Income_Category"]=oei.fit_transform(df[["Income_Category"]])
```

```python
oec=OrdinalEncoder(categories=[["Platinum","Gold","Silver","Blue"]])
```

```python
df["Card_Category"]=oec.fit_transform(df[["Card_Category"]])
```

```python
df.info()
```

```python
df
```

```python
x=df.drop('Attrition_Flag',axis=1).values
y=df['Attrition_Flag'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20,shuffle=True)
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
prediction=clf.predict(x_test)
```

```python
prediction
```

```python
confusion_matrix(y_test,prediction)
```

```python
lgbm=LGBMClassifier()
```

```python
lgbm.fit(x_train,y_train)
```

```python
pre=lgbm.predict(x_test)
```

```python
accuracy_score(y_test,prediction) #randomforest prediction
```

```python
accuracy_score(y_test,pre) #lgbm prediction
```

```python
joblib.dump(lgbm,"save_models.pkl")
```

```python
joblib.dump(clf,"save_model_clf.pkl")
```
